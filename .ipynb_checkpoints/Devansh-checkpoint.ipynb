# Data Description

---

## Features

| Feature                        | Type                | Column Name | Data Type | Notes                                                                 |
|--------------------------------|---------------------|-------------|-----------|----------------------------------------------------------------------|
| Age                            | Objective Feature   | `age`       | int (days) | Age in days                                                          |
| Height                         | Objective Feature   | `height`    | int (cm)   | Height in centimeters                                                |
| Weight                         | Objective Feature   | `weight`    | float (kg) | Weight in kilograms                                                  |
| Gender                         | Objective Feature   | `gender`    | categorical | Encoded categorical value                                            |
| Systolic blood pressure        | Examination Feature | `ap_hi`     | int        | Systolic blood pressure                                              |
| Diastolic blood pressure       | Examination Feature | `ap_lo`     | int        | Diastolic blood pressure                                             |
| Cholesterol                    | Examination Feature | `cholesterol` | categorical | 1: normal, 2: above normal, 3: well above normal                     |
| Glucose                        | Examination Feature | `gluc`      | categorical | 1: normal, 2: above normal, 3: well above normal                     |
| Smoking                        | Subjective Feature  | `smoke`     | binary     | 0: non-smoker, 1: smoker                                             |
| Alcohol intake                 | Subjective Feature  | `alco`      | binary     | 0: no alcohol, 1: consumes alcohol                                   |
| Physical activity              | Subjective Feature  | `active`    | binary     | 0: inactive, 1: physically active                                    |
| Cardiovascular disease (target)| Target Variable     | `cardio`    | binary     | 0: no disease, 1: presence of cardiovascular disease                 |



# Dataset Information

## General Overview
- Total records: **70,000**

---

## Feature Details

1. **Age (`age`)**
   - Unit: days  
   - Represents patientâ€™s age in days  

2. **Gender (`gender`)**
   - Encoded categorical values  
   - `1 = Female`, `2 = Male`  

3. **Height (`height`)**
   - Unit: centimeters (cm)  

4. **Weight (`weight`)**
   - Unit: kilograms (kg)  

5. **Systolic Blood Pressure (`ap_hi`)**
   - Integer value  
   - Represents systolic blood pressure  

6. **Diastolic Blood Pressure (`ap_lo`)**
   - Integer value  
   - Represents diastolic blood pressure  

7. **Cholesterol (`cholesterol`)**
   - Categorical values:  
     - `1 = Normal`  
     - `2 = Above normal`  
     - `3 = Well above normal`  

8. **Glucose (`gluc`)**
   - Categorical values:  
     - `1 = Normal`  
     - `2 = Above normal`  
     - `3 = Well above normal`  

9. **Smoking (`smoke`)**
   - Binary values:  
     - `0 = Non-smoker`  
     - `1 = Smoker`  

---

## Target Variable
- **Cardiovascular Disease (`cardio`)**
  - Binary values:  
    - `0 = No disease`  
    - `1 = Presence of cardiovascular disease`

```python
# Let's Start Model work...
```

```python
# Remove Outliers
df = df[(df['ap_hi'] > 80) & (df['ap_hi'] < 200)]
df = df[(df['ap_lo'] > 50) & (df['ap_lo'] < 130)]
```

```python
# Feature & Target Split
x = df.drop('cardio', axis=1)
y = df['cardio'] #answer column
```

```python
# Trainâ€“Test Split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
# Model Training
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)

# To not get this error: We use max_iter


# C:\Users\Lenovo\anaconda3\Lib\site-packages\sklearn\linear_model\_logistic.py:469: ConvergenceWarning: lbfgs failed to converge (status=1):
# STOP: TOTAL NO. of ITERATIONS REACHED LIMIT.

# Increase the number of iterations (max_iter) or scale the data as shown in:
#     https://scikit-learn.org/stable/modules/preprocessing.html
# Please also refer to the documentation for alternative solver options:
#     https://scikit-learn.org/stable/modules/linear_model.html#logistic-regression
#   n_iter_i = _check_optimize_result(
```

```python
# Model Evaluation
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

y_pred = model.predict(x_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
